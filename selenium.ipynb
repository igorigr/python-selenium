```python
import numpy as np
import easyocr
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select


servico = Service(ChromeDriverManager().install())

navegador=webdriver.Chrome(service=servico)

navegador.maximize_window()

print ("entrar na loja on line")
navegador.get("http://loja.sandbox.lojasvirtuais-br.com.br/index.php")
print("Aceitar os termos")
navegador.find_element('xpath','//*[@id="lgpd_aceito"]').click()
print ("Procurar um produto")
navegador.find_element('xpath','//*[@id="procurarPalavraTopo"]').send_keys('jogo de canecas + suporte')
print ("clicar no botao pesquisar")
navegador.find_element('xpath','//*[@id="actionPesquisaTopo"]').click()
try:
    element = WebDriverWait(navegador, 30).until(
        EC.presence_of_element_located(('xpath','//*[@id="id_Vitrine"]/div/div/div/div/div/div/input[1]'))
    )
finally:
    navegador.implicitly_wait(10)  
navegador.find_element('xpath','//*[@id="id_Vitrine"]/div/div/div/div/div/div/input[1]').click()
print ("verificar o total do pedido")
element = navegador.find_element('xpath','//*[@id="id_valor_total_1"]')

#  checkar se tem o valor do pedido total
search_text = "54,18"
if search_text in element.text:
    print(f"o elemento contem o texto {search_text}.")
else:
    print(f"o elemento nao contem o texto: {search_text}.")
    
print ("fechar o pedido")
navegador.find_element('xpath','//*[@id="id_bot_realizar_pedido"]').click() 

navegador.implicitly_wait(30)  
print ("Adicionar o cep para comprar o produto")
navegador.find_element('xpath','//*[@id="fieldCep"]').send_keys('06787500')
print ("clicar para finalizar o pedido")
navegador.find_element('xpath','//*[@id="id_form_login"]/div/div[2]/div/div[2]/div/div/div[2]/div[5]/input').click()
print ("Adicionar os dados do comprador")
#  cpf
navegador.find_element('xpath','//*[@id="fieldCnpj_Cpf"]').send_keys('93251801031')
#  nome completo
navegador.find_element('xpath','//*[@id="fieldCliente"]').send_keys('Yuri Testes')
#  data nascimento
navegador.find_element('xpath','//*[@id="fieldData_Nascimento"]').send_keys('23041980')
#  nome recebedor
navegador.find_element('xpath','//*[@id="fieldCampo_Extra_1"]').send_keys('testes a')

try:
    element = WebDriverWait(navegador, 30).until(
        EC.presence_of_element_located(('xpath','//*[@id="id_bot_continuar1"]'))
    )
finally:
    navegador.implicitly_wait(30)      
#  continuar
navegador.find_element('xpath','//*[@id="id_bot_continuar1"]').click()

#  numero residencia
navegador.find_element('xpath','//*[@id="fieldNumero"]').send_keys('178')
print ("avancar para finalizar o pedido")
navegador.find_element('xpath','//*[@id="id_bot_continuar2"]').click()
#  email
navegador.find_element('xpath','//*[@id="fieldEmail"]').send_keys('tes222222@teste3.com.br')
#  senha
navegador.find_element('xpath','//*[@id="fieldSenha"]').send_keys('123456')
#  confirmacao senha
navegador.find_element('xpath','//*[@id="fieldSenha1"]').send_keys('123456')


with open('loadkey.png', 'wb') as file:
#identify image to be captured
   l = navegador.find_element('xpath','//*[@id="loadKey"]/img')
#write file
   file.write(l.screenshot_as_png)
    
#img = navegador.find_element('xpath','//*[@id="loadKey"]/img').get_attribute("src")
#print(f"img valor: {img}.")
reader = easyocr.Reader(['pt'])
results = reader.readtext('loadkey.png')

print(results[0][1])
#  confirmacao senha
navegador.find_element('xpath','//*[@id="keyLbl"]').send_keys(results[0][1])
print ("finalizar cadastro e efetuar compra")
navegador.find_element('xpath','//*[@id="id_bot_continuar3"]').click()


ddelement= Select(navegador.find_element('xpath','//*[@id="id_fieldCod_Forma_Pagamento"]'))

# select by value 
ddelement.select_by_index(3)
navegador.find_element('xpath','//*[@id="id_bot_finalizar_pedido"]').click()


```
